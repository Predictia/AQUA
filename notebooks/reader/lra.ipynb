## Low Resolution Archive (LRA)

This shows how to access and play with the LRA and can be used as playground to check that all variables produced are correct.
We can call the LRA by selecting as sources `lra-r100-monthly` and in the future also `lra-r100-daily`. `r100` is the current LRA resolution and it is 1deg. 

Here below you can see an example from ICON and FESOM. No fixer is needed since variables have been decumulated and corrected according to the AQUA data format (GRIB-like). 
However, some inconsistency might still be present. 

We start with ICON NextGems Cycle3 global temperature timeseries

```python
from aqua import Reader
reader = Reader(model="ICON", exp="ngc3028", source="lra-r100-monthly")
data = reader.retrieve()
reader.fldmean(data['2t']).plot()
```

An example for SST in FESOM

```python
reader = Reader(model="FESOM", exp="tco2559-ng5", source="lra-r100-monthly")
data = reader.retrieve()
data['sst'][6,:,:].plot()
```

Same stuff but for IFS surface temperature

```python
reader = Reader(model="IFS", exp="tco2559-ng5", source="lra-r100-monthly")
data = reader.retrieve()
data
```
